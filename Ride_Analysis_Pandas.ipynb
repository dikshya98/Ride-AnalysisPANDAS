# Number of trips by different payment type

```python
Paymenttype_count = df.groupby('payment_type').size().reset_index(name='Total_trips')
print(Paymenttype_count)

```

```python
sns.barplot(x='payment_type',y='Total_trips',data= Paymenttype_count)
```

```python
# Step 1: Define the bins (ranges)
bins = [0, 10, 20, 30, 50, 100]  # Fare ranges you want
labels = ['0-10', '10-20', '20-30', '30-50', '50-100']  # Labels for each bin

# Step 2: Use pd.cut() to create a new column 'fare_range' based on the bins
df['fare_range'] = pd.cut(df['fare_amount'], bins=bins, labels=labels, right=False)
```

```python
Fare_Range = df.groupby('fare_range').size().reset_index(name='Trips_byrange')
print(Fare_Range)
```

```python
sns.barplot (x='fare_range',y='Trips_byrange', data= Fare_Range)
```
